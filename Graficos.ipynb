```
import matplotlib.pyplot as plt
```

```
x = [1,2,5]
y = [2,3,7]
plt.title("Meu primeiro gr√°fico em python")
#Eixo X
plt.xlabel("Eixo X")
#Eixo Y
plt.ylabel("Eixo Y")
plt.plot(x, y)
plt.show()
```

```
x = [1,2,3,4,5]
y = [2,3,7,1,0] #tamanho dos dados
titulo = "Gr√°fico de Barras"
eixox = "Eixo x"
eixoy = "Eixo Y"

#Titulo
plt.title(titulo)

#Eixos
plt.xlabel(eixox)
plt.xlabel(eixoy)
plt.bar(x,y)
plt.show()
```

```
#Comparando gr√°ficos
x1 = [1,3,5,7,9]
y1 = [2,3,7,1,0]

x2 = [2,4,6,8,10]
y2 = [5,1,3,7,4]

titulo = "Gr√°fico de barras 2"
eixox = "Eixo X"
eixoy = "Eixo Y"

#titulo
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x1,y1, label= "Grupo 1")
plt.bar(x2,y2, label ="Grupo2")
plt.legend() #mostra a label
plt.show()
```

```
x = [1,3,5,7,9]
y = [2,3,7,1,0]

titulo = "Scatterplot: gr√°fico"
eixox = "Eixo X"
eixoy = "Eixo Y"

#Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)
plt.scatter(x,y, label = "Meus pontos", color ="red", marker="^", s=100) #color, marcador e tamanho
plt.legend()
plt.plot(x,y, color = "k",linestyle=":") # Une os pontos, estilo da linha e cor
plt.show()
```

**color: cor (ver exemplos abaixo)**

**label: r√≥tulo**

**linestyle: estilo de linha (ver exemplos abaixo)**

**linewidth: largura da linha**

**marker: marcador (ver exemplos abaixo) **
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

> **CORES (color)**


'b' blue

'g' green

'r' red

'c' cyan

'm' magenta

'y' yellow

'k' black

'w' white

> **Marcadores (marker)**

'.' point marker

',' pixel marker

'o' circle marker

'v' triangle_down marker

'^' triangle_up marker

'<' triangle_left marker

'>' triangle_right marker

'1' tri_down marker

'2' tri_up marker

'3' tri_left marker

'4' tri_right marker

's' square marker

'p' pentagon marker

'*' star marker

'h' hexagon1 marker

'H' hexagon2 marker

'+' plus marker

'x' x marker

'D' diamond marker

'd' thin_diamond marker

'|' vline marker

'_' hline marker



> **Tipos de linha (linestyle)**


'-' solid line style

'--' dashed line style

'-.' dash-dot line style

':' dotted line style



```
x = [1,3,5,7,9]
y = [2,3,7,1,0]

x2 = [2,4,6,8,10]
y2 = [5,1,3,7,4]
z = [500,25,100,400,300] #determinar o tamanho das bolas

titulo = "Scatterplot: gr√°fico"
eixox = "Eixo X"
eixoy = "Eixo Y"

#Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)
plt.scatter(x,y, label = "Meus pontos", color ="red", marker=".", s=200) #color, marcador e tamanho
plt.scatter(x2,y2, label = "Pontos2", color ="b", marker=".", s=z) #color, marcador e tamanho
plt.legend()
plt.plot(x,y, color = "k",linestyle="-.") # Une os pontos, estilo da linha e cor
plt.plot(x2,y2, color = "k",linestyle="-") # Une os pontos, estilo da linha e cor
plt.show()
```
